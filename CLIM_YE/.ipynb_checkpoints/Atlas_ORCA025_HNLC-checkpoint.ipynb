# ORCA025-PISCES model-data comparison: surface maps

Make surface maps of some key biogeochemical variables from ORCA025-PISCES and compare to climatological data. To do this very basic, qualitative evaluation, use pyferret in IPython Notebook.

After installing ferretmagic and loading it, this Notebook (1) makes a list of files to be treated, (2) makes and writes out three new plotting routines (Ferret go scripts: <b>mapmod.jnl, mapdat.jnl, and mod2datgrid.jnl</b>), and (3) uses those routines in subsequent cells to make surface maps of model results vs. climatological data.

#### Install ferretmagic extension

```python
%install_ext https://raw.github.com/PBrockmann/ipython-ferretmagic/master/ferretmagic.py
%load_ext ferretmagic
```

```python
%%ferret -q
can data/all
use "http://dods.idris.fr/cgi-bin/nph-dods/rfry938/ORCA025-PIS2DIC/GRID/mask.nc"
use "http://dods.idris.fr/cgi-bin/nph-dods/rfry938/ORCA025-PIS2DIC/MBG/Analyse/CLIM_YE/ORCA025_1961_2010_10Y_NO3_surf.nc"
use "http://dods.idris.fr/cgi-bin/nph-dods/rfry938/ORCA025-PIS2DIC/MBG/Analyse/CLIM_YE/ORCA025_1961_2010_10Y_NCHL_surf.nc"
use "http://dods.idris.fr/cgi-bin/nph-dods/rfry938/ORCA025-PIS2DIC/MBG/Analyse/CLIM_YE/ORCA025_1961_2010_10Y_DCHL_surf.nc"
use "http://dods.ipsl.jussieu.fr/cgi-bin/nph-dods/ocmip/phase5/DATA/gridded/CARS2009/nitrate_cars2009.nc"
!use "/home/biomac1/blevu/DATA/CARS2009/nitrate_cars2009.nc"
use "http://dods.ipsl.jussieu.fr/cgi-bin/nph-dods/ocmip/phase5/DATA/gridded/WOA2009/nitrate_annual_1deg.nc"
!use "/home/biomac2/jpalm/WOA/nitrate_annual_1deg.nc"
use "http://dods.ipsl.jussieu.fr/cgi-bin/nph-dods/ocmip/phase5/DATA/gridded/SeaWiFS/seawifs_clim_1x1.nc"
```

#### Define MYGRID to remove time axis of tmask which is unnecessary and interferes with time_counter axis of model output

```python
%%ferret -q
def grid/x=x/y=y/z=z mygrid
set axis/name=xlon X
set axis/name=ylat Y
set axis/name=zdepth Z

set axis/name=lonCHL NO3X 
set axis/name=latCHL ETOPO60Y
```

#### Define proper lat, lon coordinates from MASK and MYGRID 

```python
%%ferret
let lon_masked=if TMASK[g=mygrid, d=1, k=1] eq 1 then nav_lon[d=1]
let lat_masked=if TMASK[g=mygrid, d=1, k=1] eq 1 then nav_lat[d=1]

set variable/units="degrees_east"/title="Longitude" lon_masked
set variable/units="degrees_north"/title="Latitude" lat_masked
```

#### Set memories and subreg for big data

```python
%%ferret
! Consider only every 4th data point in x and y directions to conserve computational time (roughly 1/4° -> 1°)
! def sym subreg=i=1:1442:4,j=1:1021:4

! Consider every data point in x and y directions (more costly; finer details may not be noticible on globap maps)
  def sym subreg=i=1:1442:4,j=1:1021:4
  set mem/siz=300
```

### HNLC maps

To use the custom palettes below, you may need to add my directory to your FER_PALETTE environment variable.
If the output below does not end in "/home/users/orr/ferret/palettes", uncomment the last 2 lines and rerun this cell.
If these palettes are still not available you can manually do this command in your .bashrc file 
or make the equivalent command in your .csrhc file (when running csh or tcsh). After testing, only the manual addition will work.

```python
%%ferret
let var1 = NO3[d=2,l=5]*16/122*1e6
let var2 = (NCHL[d=3,l=5]+DCHL[d=4,l=5])*1e6

let var1m = if var1 then var1
let var2m = if var2 then var2

let datm = if (var1m[l=5] gt 2) and (var2m[l=5] lt 0.3) then 1 else 0

let var1b=mean[d=5,k=1]
let var2b=CHL[d=7,l=@ave]
```

```python
%%ferret
! Interpolate x-y (variable NO3 (CARS) onto grid for Chl a (SeaWiFS) 
! - var1b is the NO3 from CARS
! - var2b is the Chl a from SeaWiFS
let v1_on_v2 = var1b[gx=lonCHL,gy=latCHL, k=1]
let vdummy = x[gx=lonCHL] * y[gy=latCHL]
let v1_on_v2t = reshape(v1_on_v2[k=1], vdummy)
let v2t = reshape(var2b, v1_on_v2t)
```

```python
%%ferret -q -s 800,400
!%%ferret -q -p -f surfmap_ORCA025-PISCES_NO3.pdf

let datd = if (v1_on_v2t gt 2) and (v2t lt 0.3) then 2 else 0

go page_new 1 3 5 95 0 100 0 0
set win/clear
go boldf_plattner

! 1) Model
set view 1
go margins_set 15 8 8 10
def sym opal nair1_cmyk
go mapmod datm[($subreg)] lon_masked[($subreg)] lat_masked[($subreg)] no Latitude "/lev=(0,1,0.5)/pal=($opal)"

! 2) Data
  set view 2
  go margins_set 8 15 8 10
  go mapdat datd no Latitude "/lev=(0,2,1)/pal=($opal)"

! 3) Model - Data difference
  set view 3
  go margins_set 1 22 8 10
  def sym opal light_centered
! Regrid model to data grid
  go mod2datgrid datm datd lon_masked lat_masked
! Compute model - data difference (on data grid) 
! -> reshape command below works around bug in Ferret 6.9 (should not be needed; without it, cannot make map w/ right LON limits)
  let diffmd = reshape(datm_reg - datd, datd)
! Make map
  go mapdat diffmd Longitude Latitude "/lev=(-2,2,1)/pal=($opal)"
```

#### Fig. 15. HNLC area maps: (top) model 10 years mean (2001-2010), (middle) CARS&SeaWiFS climtology, and (bottom) model-data difference. Difference model-data with red when neither model or data is a HNLC area, yellow when only model diagnostic HNLC area, light blue when it is only data and dark blue where both model and data are agree.  

We considere here the limit of HNLC region to be areas where CHL is below 0.3 when NO3 is more than 2µM roughly according to the definition of Sarmiento and Gruber 2006.
The model sems overestimates the extenction of HNLC areas, inparticular the North Pacific and a bit the Southern Atlantic. This due to the underestimation of Chlorophyll a in the north Pacific by the model by comparison with SeaWiFS climatology. In contrast the model almost never under estimate the areas extentin of HNLC if we compare with the calcul with CARS and SeaWiFS (1997-2005) climatology.
