# data comparison: surface zonal means

Synopsis: Makes surface zonal means of Mixed Layer Depth (MLD) from different climatological data and compare them.

Procedure: (1) Install and load ferretmagic, (2) read data files and (3) use them to plot surface zonal means.

Plots (see below): zonal mean observed (1) MLDave, (2) MLDmin, and (3) MLDmax.

#### Install & load ferretmagic extension

```python
%install_ext https://raw.github.com/PBrockmann/ipython-ferretmagic/master/ferretmagic.py
%load_ext ferretmagic
```

#### Get MODEL OUTPUT from prodn server on IDRIS http://prodn.idris.fr/thredds/dodsC/IDRISPUBFS/rfry451/ORCA025-PIS2DIC/catalog.html

#### Get DATA from DODS server on ciclad, i.e. http://dods.ipsl.jussieu.fr/cgi-bin/nph-dods/ocmip/phase5/DATA/gridded

```python
%%ferret -q
  can data/all
! CARS MLD criteria DT 0.4degC or DS 0.3
! use "http://dods.ipsl.jussieu.fr/cgi-bin/nph-dods/ocmip/phase5/DATA/gridded/CARS2009/mld_cars2009a.nc"
  use "/home/biomac1/blevu/DATA/CARS2009/mld_monthly_cars2009a.nc"
! Boyer_Montegut criteria DT 0.2
  use "http://dods.ipsl.jussieu.fr/cgi-bin/nph-dods/ocmip/phase5/DATA/gridded/de_Boyer_Montegut/mld_DT02.nc"
! Boyer_Montegut criteria density 0.03 kg/m3
  use "http://dods.ipsl.jussieu.fr/cgi-bin/nph-dods/ocmip/phase5/DATA/gridded/de_Boyer_Montegut/mld_DR003.nc"
```

### Plot 1: minimum MLD zonal means (CARS2009 and Boyer Montegut)

#### Prepare zonal mean MLD data (put on same regular grid)

```python
%%ferret -q
! Get basin masks and reference grid (World Ocean Atlas, 1° x 1°)
  def sym readbmask "notyet"
  go choosebasin glo

! DATA output:
  let var1=mld[d=1,l=@min]
  let datd1 = if var1 then var1

  set var/bad=1.e+09 mld[d=2]
  let datd2=mld[d=2,l=@min]

  set var/bad=1.e+09 mld[d=3]
  let datd3=mld[d=3,l=@min]

! Specify reference field (the grid onto which both the model & the data will be placed)
  let ref = mask[d=WOA2001_grid.nc, k=1]

! Regrid data to reference grid (World Ocean Atlas) - produces "datd_refgrid"
  go dat2refgrid datd1 ref
  let datd1_ref = datd1_refgrid
  go dat2refgrid datd2 ref
!  let datd2_ref = reshape(datd2_refgrid, datd1_ref)
  let datd2_ref = datd2_refgrid
  go dat2refgrid datd3 ref
!  let datd3_ref = reshape(datd3_refgrid, datd1_ref)
  let datd3_ref = datd3_refgrid
```

#### Plot zonal means (Global, Atlantic, Indian, Pacific)

```python
%%ferret -q -s 800,400
! Choose color palette & define line patterns:
  def sym opal nair1_cmyk
! def sym opal light_centered

  def sym dot "(.025,.025,.025,.025)"
  def sym dotwide "(.045,.045,.045,.045)"
  def sym dashdot "(.1,.05,.025,.05)"
  def sym dotwider "(.045,.1,.045,.1)"

! Make 3 plots:
! -------------
  go page_new -l 2 2 5 95 10 90 0 0
  set win/clear
  go boldf_plattner

! Refine plots with a some ferret (PPL) setup commands
  ppl color 3    0, 55,   0    !dark green instead of default green (too light)
  ppl axlabp -1, -1            !labeled axes on left and bottom
  ppl axnmtc 2, 4              !number of minor ticks on x and y axes
  ppl axlint 1, 1              !numer  of major ticks to label (1 = every, 2 = every other)
  def sym yformat (i5)

! Specify y axis limits 
  def sym ymin -10 ; def sym ymax 100 ; def sym yinc 20
  def sym ylims ($ymin):($ymax):($yinc)
  let ysubtitle = `($ymin)` + 0.9*`($ymax) - ($ymin)`

! 1) Global
  set view 1
  go margins_set 15 8 13 5
  go zmsurf datd1_ref glo no "MLD (m)"  -90 +90 "/nolab/set/col=red/thick=2/hlim=90s:90n:30/vlim=($ylims)/line" firstplt
  go zmsurf datd2_ref glo no no -90 +90 "/nolab/over/col=blue/thick=2/dash=($dotwide)  " over
  go zmsurf datd3_ref glo no no -90 +90 "/nolab/over/col=black/thick=2/dash=($dotwide)  " over
  label 87 `ysubtitle` +1 0 .20 "Global"

! 2) Atlantic
  set view 2
  go margins_set 15 8 18 0
  go zmsurf datd1_ref atlarc no no -90 +90 "/nolab/set/col=red/thick=2/hlim=90s:90n:30/vlim=($ylims)/line" firstplt
  go zmsurf datd2_ref atlarc no no -90 +90 "/nolab/over/col=blue/thick=2/dash=($dotwide)  " over
  go zmsurf datd3_ref atlarc no no -90 +90 "/nolab/over/col=black/thick=2/dash=($dotwide)  " over
  label 87 `ysubtitle` +1 0 .20 "Atlantic"

! 3) Indian
  set view 3
  go margins_set 15 8 13 5
  go zmsurf datd1_ref ind no "MLD (m)"  -90 +90 "/nolab/set/col=red/thick=2/hlim=90s:90n:30/vlim=($ylims)/line" firstplt
  go zmsurf datd2_ref ind no no -90 +90 "/nolab/over/col=blue/thick=2/dash=($dotwide)  " over
  go zmsurf datd3_ref ind no no -90 +90 "/nolab/over/col=black/thick=2/dash=($dotwide)  " over
  label 87 `ysubtitle` +1 0 .20 "Indian"

! 4) Pacific
  set view 4
  go margins_set 15 8 18 0
  go zmsurf datd1_ref pac no no -90 +90 "/nolab/set/col=red/thick=2/hlim=90s:90n:30/vlim=($ylims)/line" firstplt
  go zmsurf datd2_ref pac no no -90 +90 "/nolab/over/col=blue/thick=2/dash=($dotwide)  " over
  go zmsurf datd3_ref pac no no -90 +90 "/nolab/over/col=black/thick=2/dash=($dotwide)  " over
  label 87 `ysubtitle` +1 0 .20 "Pacific"
```

#### Fig. 1. Surface zonal means of minimum MLD for the CARS data (red), Boyer Montegut with T criteria (blue) or with density criteria (black)  over the global ocean as well as the Atlantic, Indian, and Pacific basins.

The MLD min for CARS2009 is a between the 2 MLD of Boyer_MOntegut in low latitude and a bit stronger in high latitude. The criteria for CARS gives a proper value for each latitude not like Boyer Montegut data for unknown reason.

### Plot 2: average MLD means (CARS2009 and Boyer Montegut)

#### Prepare zonal mean MLD data (put on same regular grid)

```python
%%ferret -q
! Get basin masks and reference grid (World Ocean Atlas, 1° x 1°)
  def sym readbmask "notyet"
  go choosebasin glo

! DATA output:
  let var1=mld[d=1,l=@ave]
  let datd1 = if var1 then var1

! set var/bad=1.e+09 mld[d=2]
  let datd2=mld[d=2,l=@ave]

!  set var/bad=1.e+09 mld[d=3]
  let datd3=mld[d=3,l=@ave]

! Specify reference field (the grid onto which both the model & the data will be placed)
  let ref = mask[d=WOA2001_grid.nc, k=1]

! Regrid data to reference grid (World Ocean Atlas) - produces "datd_refgrid"
  go dat2refgrid datd1 ref
  let datd1_ref = datd1_refgrid
  go dat2refgrid datd2 ref
  let datd2_ref = reshape(datd2_refgrid, datd1_ref)
  go dat2refgrid datd3 ref
  let datd3_ref = reshape(datd3_refgrid, datd1_ref)
```
