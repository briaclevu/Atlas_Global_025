Notebook to visualze global ORCA025-PISCES output using PyFerret

#### First thing is to install the ferret extension

```python
%install_ext https://raw.github.com/PBrockmann/ipython-ferretmagic/master/ferretmagic.py
```

#### Then load the extension

```python
%load_ext ferretmagic
```

#### Build a list of files from http://dods.idris.fr/rfry451/ORCA025-PIS2DIC/ in remote access and open it

```python
%%ferret
use "http://prodn.idris.fr/thredds/dodsC/IDRISPUBFS/rfry451/ORCA025-PIS2DIC/GRID/mask.nc"
use "http://prodn.idris.fr/thredds/dodsC/IDRISPUBFS/rfry451/ORCA025-PIS2DIC/OCE/Output/YE/ORCA025-PIS2DIC_20000101_20001231_1Y_grid_T.nc"
use "http://prodn.idris.fr/thredds/dodsC/IDRISPUBFS/rfry451/ORCA025-PIS2DIC/MBG/Output/YE/ORCA025-PIS2DIC_20000101_20001231_1Y_ptrc_T.nc"
sh d
use "/home/biomac1/blevu/DATA/CARS2009/temperature_cars2009a.nc"
use "/home/biomac2/jpalm/WOA/temperature_monthly_1deg.nc"
use "/home/biomac1/blevu/DATA/CARS2009/salinity_cars2009a.nc"
use "/home/biomac2/jpalm/WOA/salinity_monthly_1deg.nc"
use "/home/biomac1/blevu/DATA/CARS2009/oxygen_cars2009.nc"
use "/home/biomac2/jpalm/WOA/dissolved_oxygen_annual_1deg.nc"
```

#### Define MYGRID to remove time axis of tmask which is in contradiction with time_counter axis of modele output

```python
%ferret_run 'def grid/x=x/y=y/z=z mygrid'
```

#### Define proper coordinates from MASK and MYGRID 

```python
%%ferret
let lon_masked=if TMASK[g=mygrid, d=1, k=1] eq 1 then nav_lon[d=1]
let lat_masked=if TMASK[g=mygrid, d=1, k=1] eq 1 then nav_lat[d=1]

set variable/units="degrees_east"/title="Longitude" lon_masked
set variable/units="degrees_north"/title="Latitude" lat_masked
```

#### Set memories and subreg for big data

```python
%%ferret
! set mem/siz=300
  def sym subreg=i=1:1442:4,j=1:1021:4
```

#### Plot masked Temperature at the surface in 2000 model & CARS (or WOA)

```python
%%ferret -q -s 800,400
let var=votemper[d=2]
let myvar=if TMASK[d=1,g=mygrid] eq 1 then var

set win/size=1/new/aspect=1.2
set viewport upper
shade/hlim=0:360:30/vlim=-90:90:30/levels=(-2,30,2)/modulo myvar[k=2,l=1,($subreg)], lon_masked[($subreg)], lat_masked[($subreg)]
go land
go fland

set viewport lower
shade/hlim=0:360:30/vlim=-90:90:30/levels=(-2,30,2) mean[d=4,k=3]
!shade/hlim=0:360:30/vlim=-90:90:30/levels=(-2,30,2) T_AN[d=5,k=3]
go land
go fland
```

#### Plot masked Salinity at the surface in 2000 model & CARS (or WOA)

```python
%%ferret -q -s 800,400
let var=vosaline[d=2]
let myvar=if TMASK[d=1,g=mygrid] eq 1 then var

set win/size=1/new/aspect=1.2
set viewport upper
shade/hlim=0:360:30/vlim=-90:90:30/levels=(30,40,0.5)/modulo myvar[k=2,l=1,($subreg)], lon_masked[($subreg)], lat_masked[($subreg)]
go land
go fland

set viewport lower
shade/hlim=0:360:30/vlim=-90:90:30/levels=(30,40,0.5) mean[d=6,k=3]
!shade/hlim=0:360:30/vlim=-90:90:30/levels=(30,40,0.5) S_AN[d=7,k=2]
go land
go fland
```

#### Masked O2 at the surface in 2000 model & CARS (or WOA)

```python
%%ferret -q -s 800,400
let var=O2[d=3]*1e6
let myvar=if TMASK[d=1,g=mygrid] eq 1 then var
let O2mol=mean[d=8]*44.659

set win/size=1/new/aspect=1.2
set viewport upper
shade/hlim=0:360:30/vlim=-90:90:30/levels=(100,500,20)/modulo myvar[k=2,l=1,($subreg)], lon_masked[($subreg)], lat_masked[($subreg)]
go land
go fland

set viewport lower
shade/hlim=0:360:30/vlim=-90:90:30/levels=(100,500,20) O2mol[k=3]
go land
go fland
```

#### Masked AOU at the surface in 2000 model & GLODAP

```python
%%ferret -q -s 800,400

! solubility from Weiss (1970) Deep Sea Research V17(4): 721-735

let T1 = (votemper[d=2] + 273.15) / 100
let S1 = vosaline[d=2]
let OSAT1 = -177.7888 + 255.5907 / T1 + 146.4813 * log(T1) - 22.2040 * T1
let OSAT2 = OSAT1 + S1 * (-0.037362 + T1 * (0.016504 - 0.0020564 * T1))
let OSAT3 = 10^(OSAT2)

let OSAT = OSAT3 * 1000 / 22.392
let var = OSAT - O2[d=3]*1e6/1.028
let myvar=if TMASK[d=1,g=mygrid] eq 1 then var

let T2 = (MEAN[d=4] + 273.15) / 100
let S2 = MEAN[d=6]
let OSAT1b = -177.7888 + 255.5907 / T2 + 146.4813 * log(T2) - 22.2040 * T2
let OSAT2b = OSAT1b + S2 * (-0.037362 + T2 * (0.016504 - 0.0020564 * T2))
let OSAT3b = 10^(OSAT2b)

let OSATb = OSAT3b * 1000 / 22.392
let AOU = OSATb - O2mol/1.028

set win/size=1/new/aspect=1.2
set viewport upper
shade/hlim=0:360:30/vlim=-90:90:30/levels=(-450,-150,20)/modulo myvar[k=2,l=1,($subreg)], lon_masked[($subreg)], lat_masked[($subreg)]
go land
go fland

set viewport lower
shade/hlim=0:360:30/vlim=-90:90:30/levels=(-450,-150,20) AOU[k=3]
go land
go fland
```
