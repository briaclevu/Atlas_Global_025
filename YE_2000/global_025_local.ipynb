```python
%install_ext https://raw.github.com/PBrockmann/ipython-ferretmagic/master/ferretmagic.py
```

#### Then load the extension

```python
%load_ext ferretmagic
```

#### Build a list of files from http://dods.idris.fr/rfry451/ORCA025-PIS2DIC/ in remote access and open it

```python
%%ferret
use "/home/biomac1/blevu/GLOBAL_025/mesh_zgr.nc"
use "/home/biomac1/blevu/GLOBAL_025/mask.nc"
!use "http://dods.idris.fr/cgi-bin/nph-dods/rfry451/ORCA025-PIS2DIC/MBG/Output/MO/ORCA025-PIS2DIC_20000101_20001231_1M_PO4.nc"
use "/home/biomac1/blevu/GLOBAL_025/ORCA025-PIS2DIC_20000101_20001231_1M_PO4.nc"
use "/home/biomac1/blevu/DATA/CARS2009/phosphate_cars2009.nc"
use "/home/biomac2/jpalm/WOA/phosphate_annual_1deg.nc"
sh d
```

#### Reshape mask to remove T axis which is in contradiction with time_counter axis of modele output

```python
%%ferret
  let var0 = x[gx=PO4[d=3]]*y[gy=PO4[d=3]]*z[gz=PO4[d=3]]
  let my_mask=if reshape(tmask[d=2],var0) eq 1 then 1

! let PO4m = PO4*my_mask
! let PO4m = if PO4 gt 0 then PO4
  let PO4m = if my_mask eq 1 then PO4[d=3]
```

#### Set memories and subreg for big data

```python
%%ferret

  set mem/siz=300

! set reg/i=1:1442:5/j=1:1021:5

! def sym subreg gi=1:1442:1,gj=1:1021:1
  def sym subreg gi=1:1442:5,gj=1:1021:5

! sho sym ($subreg)
```

#### Plot surface PO4 in 2000 model & CARS & WOA

```python
%%ferret
! shade/hlim=120:480:30/vlim=-90:90:30/mod PO4m[k=1,l=@ave], nav_lon, nav_lat
! shade/nolab/hlim=120:480:30/vlim=-90:90:30/mod PO4m[k=1,l=@ave,($subreg)], nav_lon[($subreg)], nav_lat[($subreg)]

!%ferret_run -q -s 800,400 'shade/hlim=120:480:30/vlim=-90:90:30/mod PO4m[k=1,l=@ave,($subreg)], nav_lon[d=3,($subreg)], nav_lat[d=3,($subreg)]' % locals()

!%ferret_run -q -s 800,400 'shade PO4m[k=1,l=@ave], nav_lon, nav_lat' % locals()
!%ferret_run -q -s 800,400 'shade PO4[k=1,l=@ave]' % locals()
```

```python
%ferret_run -q -s 800,400 'shade/hlim=0:360:30/vlim=-90:90:30/mod/levels=(0,2,0.2) PO4m[k=2,l=@ave,($subreg)]/122*1e6, nav_lon[d=3,($subreg)], nav_lat[d=3,($subreg)]; go land; go fland' % locals()
%ferret_run -q -s 800,400 'shade/levels=(0,2,0.2) mean[d=4,k=3]; go land; go fland'
%ferret_run -q -s 800,400 'shade/levels=(0,2,0.2) P_AN[d=5,k=2]; go land; go fland'
```
