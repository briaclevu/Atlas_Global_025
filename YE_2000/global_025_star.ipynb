---

#### First thing is to install the ferret extension

```python
%install_ext https://raw.github.com/PBrockmann/ipython-ferretmagic/master/ferretmagic.py
```

#### Then load the extension

```python
%load_ext ferretmagic
```

#### Build a list of files from http://dods.idris.fr/rfry451/ORCA025-PIS2DIC/ in remote access and open it

```python
%%ferret
use "http://prodn.idris.fr/thredds/dodsC/IDRISPUBFS/rfry451/ORCA025-PIS2DIC/GRID/mask.nc"
use "http://prodn.idris.fr/thredds/dodsC/IDRISPUBFS/rfry451/ORCA025-PIS2DIC/MBG/Output/YE/ORCA025-PIS2DIC_20000101_20001231_1Y_ptrc_T.nc"
sh d
use "/home/biomac1/blevu/DATA/CARS2009/phosphate_cars2009.nc"
use "/home/biomac2/jpalm/WOA/phosphate_annual_1deg.nc"
use "/home/biomac1/blevu/DATA/CARS2009/nitrate_cars2009.nc"
use "/home/biomac2/jpalm/WOA/nitrate_annual_1deg.nc"
use "/home/biomac1/blevu/DATA/CARS2009/silicate_cars2009.nc"
use "/home/biomac2/jpalm/WOA/silicate_annual_1deg.nc"
use "/home/biomac1/blevu/DATA/GLODAP/tco2_ann_Key.cdf"    
use "/home/biomac1/blevu/DATA/GLODAP/talk_ann_Key.cdf"
```

#### Define MYGRID to remove time axis of tmask which is in contradiction with time_counter axis of modele output

```python
%ferret_run 'def grid/x=x/y=y/z=z mygrid'
```

#### Define proper coordinates from MASK and MYGRID 

```python
%%ferret
let lon_masked=if TMASK[g=mygrid, d=1, k=1] eq 1 then nav_lon[d=1]
let lat_masked=if TMASK[g=mygrid, d=1, k=1] eq 1 then nav_lat[d=1]

set variable/units="degrees_east"/title="Longitude" lon_masked
set variable/units="degrees_north"/title="Latitude" lat_masked
```

#### Set memories and subreg for big data

```python
%%ferret
! set mem/siz=300
  def sym subreg=i=1:1442:4,j=1:1021:4
```

#### Plot N-star in 2000 model & CARS (or WOA)

```python
%%ferret -q -s 800,400
let var=(NO3[d=2]-PO4[d=2])*0.87*16/122*1e6+2.523
let myvar=if TMASK[d=1,g=mygrid] eq 1 then var
let NstarC=(MEAN[d=5]-16*MEAN[d=3]+2.9)*0.87
let NstarW=(N_AN[d=6]-16*P_AN[d=4]+2.9)*0.87

set win/size=1/new/aspect=1.2
set viewport upper
shade/hlim=0:360:30/vlim=-90:90:30/modulo/levels=(-10,5,1) myvar[k=2,l=1,($subreg)], lon_masked[($subreg)], lat_masked[($subreg)]
!shade/hlim=0:360:30/vlim=-90:90:30/modulo/levels=(0,2,.2) myvar[k=2,l=1,($subreg)], lon_masked[($subreg)], lat_masked[($subreg)]
go land
go fland

set viewport lower
shade/hlim=0:360:30/vlim=-90:90:30/levels=(-10,5,1) NstarC[k=3]
!shade/hlim=0:360:30/vlim=-90:90:30/levels=(0,2,.2) mean[d=3,k=3]
!shade/hlim=0:360:30/vlim=-90:90:30/levels=(0,2,.2) P_AN[d=4,k=3]
go land
go fland

!set viewport lower
!shade/hlim=0:360:30/vlim=-90:90:30/levels=(-10,5,1) NstarW[k=2]
!shade/hlim=0:360:30/vlim=-90:90:30/levels=(0,2,.2) mean[d=3,k=3]
!shade/hlim=0:360:30/vlim=-90:90:30/levels=(0,2,.2) P_AN[d=4,k=3]
!go land
!go fland
```

#### Plot masked Si-star at the surface in 2000 model & CARS (or WOA)

```python
%%ferret -q -s 800,400
let var=(Si[d=2]-NO3[d=2]*16/122)*1e6
let myvar=if TMASK[d=1,g=mygrid] eq 1 then var
let IstarC=MEAN[d=7]-MEAN[d=5]
let IstarW=I_AN[d=8]-N_AN[d=6]

set win/size=1/new/aspect=1.2
set viewport upper
shade/hlim=0:360:30/vlim=-90:90:30/levels=(-10,20,2)/modulo myvar[k=2,l=1,($subreg)], lon_masked[($subreg)], lat_masked[($subreg)]
go land
go fland

set viewport lower
shade/hlim=0:360:30/vlim=-90:90:30/levels=(-10,20,2) IstarC[k=3]
!shade/hlim=0:360:30/vlim=-90:90:30/levels=(0,30,2) N_AN[d=6,k=2]
go land
go fland
```

#### Masked TALK-DIC at the surface in 2000 model & GLODAP

```python
%%ferret -q -s 800,400
let var=(ALKALINI2[d=2]-DIC2[d=2])*1e6
let myvar=if TMASK[d=1,g=mygrid] eq 1 then var
let DAC=TALK[d=10]-TCO2[d=9]


set win/size=1/new/aspect=1.2
set viewport upper
shade/hlim=-180:180:30/vlim=-90:90:30/levels=(-100,500,50)/modulo myvar[k=2,l=1,($subreg)], lon_masked[($subreg)], lat_masked[($subreg)]
go land
go fland

set viewport lower
shade/hlim=-180:180:30/vlim=-90:90:30/levels=(-100,500,50) DAC[k=2]
go land
go fland
```

--

* Pacific is N-limited instead of Si limited at the equator and P-limited everywhere else but Bering sea.
* Model miss the Si-limitation 40-50 S.
* TALK-DIC is ok in the model
